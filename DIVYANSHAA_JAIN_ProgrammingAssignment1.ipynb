```python
# Problem 1: GCD using Euclid Algorithm

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return abs(a)
```

```python
print(gcd(12, 8))
print(gcd(48, 18))
print(gcd(-12, 8))
```

```python
#Euler Limit Approximation

def euler_limit(n):
    return (1 + 1/n)**n
```

```python
print(euler_limit(10))
print(euler_limit(100))
print(euler_limit(1000000000000000000000000000000000))
```

```python
# When we call euler_limit for really huge values, the limit evaluates to 1
# This is because when n is very large, 1/n becomes very small with a lot of digits after the decimal point
# Python cannot represent these digits accurately, and the limit results in 1 as the final answer
```

```python
# Factorial Function

def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result
```

```python
print(factorial(5))
```

```python
# Euler Sum

def euler_sum(n):
    total = 0
    for k in range(n):
        total += 1 / factorial(k)
    return total
```

```python
print(euler_sum(1))
print(euler_sum(4))
print(euler_sum(10))
```

```python
#Euler Approx until epsilon

import math

def euler_approx(epsilon):
    total = 0
    k = 0

    while abs(math.e - total) > epsilon:
        total += 1 / factorial(k)
        k += 1

    return total
```

```python
print(euler_approx(0.01))
print(euler_approx(0.0001))
```

```python
# Error Printer

import math

def print_euler_sum_error(n):
    for k in range(1, n+1):
        error = euler_sum(k) - math.e
        print(error)
```

```python
print_euler_sum_error(5)
```

```python
# Euler Limit Error Printer

def print_euler_lim_error(n):
    for k in range(1, n+1):
        error = euler_limit(k) - math.e
        print(error)
```

```python
#Power of 2

def is_power_of_2(n):
    while n > 1:
        if n % 2 != 0:
            return False
        n = n // 2
    return True
```

```python
print(is_power_of_2(8))
print(is_power_of_2(12))
```

```python
#General Power

def is_power(b, n):
    while n > 1:
        if n % b != 0:
            return False
        n = n // b
    return True
```

```python
print(is_power(3, 27))
print(is_power(2, 20))
```
